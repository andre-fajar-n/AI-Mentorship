```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Data Loading

```python
data = pd.read_csv('datasets/boston_housing.csv')
data
```

```python
data.info()
```

## Data Visualization

```python
column = len(data.columns)
sub_plot_x = 4
sub_plot_y = column//sub_plot_x
if column % sub_plot_x != 0:
    sub_plot_y += 1
```

```python
plt.figure(figsize=(20,20))
for num_plot in range(1, sub_plot_x*sub_plot_y+1):
    if num_plot <= len(data.columns):
        column_name = data.columns[num_plot-1]
        plt.subplot(sub_plot_x, sub_plot_y, num_plot)
        sns.boxplot(data=data[column_name])
        plt.title(column_name)
```

```python
plt.figure(figsize=(20,20))
for num_plot in range(1, sub_plot_x*sub_plot_y+1):
    if num_plot <= len(data.columns):
        column_name = data.columns[num_plot-1]
        plt.subplot(sub_plot_x, sub_plot_y, num_plot)
        sns.distplot(data[column_name])
        plt.title(column_name)
```

### Split Data

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data = train_test_split(data, test_size=0.2, random_state=12)
```

```python
train_label = train_data['medv']
train_feature = train_data.drop(['medv'],axis=1)
test_label = test_data['medv']
test_feature = test_data.drop(['medv'],axis=1)
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
```

```python
## Data Training
lr_model.fit(X=train_feature, y=train_label)
```

```python
from sklearn.metrics import r2_score
```

```python
accuracy_score = lr_model.score(test_feature,test_label)
```

```python
print(accuracy_score)
```

```python
medv_predict = lr_model.predict(test_feature)
```

```python
medv_predict_df = pd.DataFrame(data={"medv_predict":medv_predict})
medv_predict_df
```

```python
test_data.loc[:, 'medv_predict'] = medv_predict
```

```python
x = np.arange(len(medv_predict))
plt.figure(figsize=(20,10))
plt.plot(x, test_label, label='Test Value')
plt.plot(x, medv_predict, label='Predict Value')
plt.legend()
```
