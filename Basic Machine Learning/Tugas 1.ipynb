```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_csv('datasets/titanic.csv')
raw_data
```

## Single Variable

```python
raw_data['Sex'].hist()
plt.show()
```

```python
raw_data['Survived'].hist()
plt.show()
```

```python
raw_data['Age'].hist()
plt.show()
```

```python
raw_data['Pclass'].hist()
plt.show()
```

## Two Variable

```python
sns.boxplot(data=raw_data, x='Survived', y='Age')
plt.show()
```

```python
sns.boxplot(data=raw_data, x='Pclass', y='Age')
plt.show()
```

```python
sns.boxplot(data=raw_data, x='Pclass', y='Fare')
plt.show()
```

## Multi Variable

```python
sns.pairplot(data=raw_data, hue='Sex')
plt.show()
```

```python
sns.pairplot(data=raw_data, hue='Survived')
plt.show()
```

```python
sns.heatmap(data=raw_data.corr(), annot=True)
plt.show()
```
